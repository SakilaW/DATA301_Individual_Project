**Loading in data for four star clusters: *NGC 2437*, *NGC 6819*, *NGC 7789*, and *Trumpler 5***

```python
## Loading in Trumpler 5 star cluster data
trumpler_5 = pd.read_csv('../Data/trumpler_5.csv')
trumpler_5.shape
```

<br>

```python
## Understanding class balance (or imbalance) in each cluster dataset
clusters = {'ngc_2437': ngc_2437, 'ngc_6819': ngc_6819, 'ngc_7789': ngc_7789, 'trumpler_5': trumpler_5}
value_counts = {}

for name, df in clusters.items():
    counts = df['Cluster'].value_counts()
    value_counts[name] = counts

for df_name, counts in value_counts.items():
    print(f'Value Counts for {df_name}:')
    print(counts)
    print()
```

<br>

```python
## Number of missing values in each column for each of the four star clusters
missing_values_ngc_2437 = ngc_2437.isnull().sum()
missing_values_ngc_6819 = ngc_6819.isnull().sum()
missing_values_ngc_7789 = ngc_7789.isnull().sum()
missing_values_trumpler_5 = trumpler_5.isnull().sum()

missing_values_df = pd.DataFrame({
    'Column Name': missing_values_ngc_2437.index,
    'Missing Values NGC 2437': missing_values_ngc_2437.values,
    'Missing Values NGC 6819': missing_values_ngc_6819.values,
    'Missing Values NGC 7789': missing_values_ngc_7789.values,
    'Missing Values Trumpler 5': missing_values_trumpler_5.values
})

missing_values_df.iloc[0:12, :]
```

**Feature Correlations, for each of the four clusters**

```python
## Feature Correlations for NGC 2437
ngc2437_CM = ngc_2437.drop(columns = ['source_id']).corr()
mask = np.triu(np.ones_like(ngc2437_CM, dtype = bool))
plt.figure(figsize = (8, 6))
sns.heatmap(ngc2437_CM, annot = True, cmap = 'coolwarm', fmt = ".2f", mask = mask)
plt.title("NGC 2437 Region: Feature Correlation Triangle Plot")
plt.show()
```

```python
## Feature Correlations for NGC 6819
ngc6819_CM = ngc_6819.drop(columns = ['source_id']).corr()
mask = np.triu(np.ones_like(ngc6819_CM, dtype = bool))
plt.figure(figsize = (8, 6))
sns.heatmap(ngc6819_CM, annot = True, cmap = 'coolwarm', fmt = ".2f", mask = mask)
plt.title("NGC 6819: Feature Correlation Triangle Plot")
plt.show()
```

```python
## Feature Correlations for NGC 7789
ngc7789_CM = ngc_7789.drop(columns = ['source_id']).corr()
mask = np.triu(np.ones_like(ngc7789_CM, dtype = bool))
plt.figure(figsize = (8, 6))
sns.heatmap(ngc7789_CM, annot = True, cmap = 'coolwarm', fmt = ".2f", mask = mask)
plt.title("NGC 7789 Region: Feature Correlation Triangle Plot")
plt.show()
```

```python
## Feature Correlations for Trumpler 5
trumpler5_CM = trumpler_5.drop(columns = ['source_id']).corr()
mask = np.triu(np.ones_like(trumpler5_CM, dtype = bool))
plt.figure(figsize = (8, 6))
sns.heatmap(trumpler5_CM, annot = True, cmap = 'coolwarm', fmt = ".2f", mask = mask)
plt.title("Trumpler 5: Feature Correlation Triangle Plot")
plt.show()
```

**Declination (DEC) vs. Right Ascension (RA), for each of the four clusters**

```python
## DEC vs. RA for NGC 2437 cluster
plt.figure(figsize = (8, 8))
sns.scatterplot(x = 'ra', y = 'dec', hue = 'parallax', data = ngc_2437, palette = 'rocket', alpha = 0.75, legend = False)

plt.xlabel('Right Ascension [degrees]')
plt.ylabel('Declination [degrees]')
plt.title('NGC 2437: Right Ascension vs. Declination, by Parallax')

sm = plt.cm.ScalarMappable(cmap = 'rocket', norm = plt.Normalize(vmin = ngc_2437['parallax'].min(), vmax = ngc_2437['parallax'].max()))
sm.set_array([])  
cbar = plt.colorbar(sm, label = 'Parallax [mas/yr]')
plt.show()
```

```python
## DEC vs. RA for NGC 6819 cluster
plt.figure(figsize = (8, 8))
sns.scatterplot(x = 'ra', y = 'dec', hue = 'parallax', data = ngc_6819, palette = 'rocket', alpha = 0.75, legend = False)

plt.xlabel('Right Ascension [degrees]')
plt.ylabel('Declination [degrees]')
plt.title('NGC 6819 Region: Right Ascension vs. Declination, by Parallax')

sm = plt.cm.ScalarMappable(cmap = 'rocket', norm = plt.Normalize(vmin = ngc_6819['parallax'].min(), vmax = ngc_6819['parallax'].max()))
sm.set_array([])  
cbar = plt.colorbar(sm, label = 'Parallax [mas/yr]')
plt.show()
```

```python
## DEC vs. RA for NGC 7789 cluster
filtered_ngc_7789 = ngc_7789[ngc_7789['ra'] > 350]

plt.figure(figsize = (8, 8))
sns.scatterplot(x = 'ra', y = 'dec', hue = 'parallax', data = filtered_ngc_7789, palette = 'rocket', alpha = 0.75, legend = False)

plt.xlabel('Right Ascension [degrees]')
plt.ylabel('Declination [degrees]')
plt.title('NGC 7789 Region: Scatter Plot of Right Ascension vs. Declination, by Parallax')

sm = plt.cm.ScalarMappable(cmap = 'rocket', norm = plt.Normalize(vmin = filtered_ngc_7789['parallax'].min(), vmax = filtered_ngc_7789['parallax'].max()))
sm.set_array([])  

cbar = plt.colorbar(sm, label = 'Parallax [mas/yr]')
plt.show()
```

```python
## NGC 2437 Colour-Magnitude Diagram
ngc_2437['absolute_magnitude'] = ngc_2437['phot_g_mean_mag'] - 5 * np.log10(ngc_2437['parallax']) + 5

plt.figure(figsize = (8, 6))
plt.scatter(ngc_2437['bp_rp'], ngc_2437['absolute_magnitude'], s = 10, c = ngc_2437['bp_rp'], cmap = 'coolwarm', marker = 'o')
plt.xlabel('Color (BP - RP)')
plt.ylabel('Absolute Magnitude (M)')
plt.gca().invert_yaxis() 
plt.title('Colour-Magnitude Diagram for NGC 2437 Cluster')
plt.show()
```

```python
## NGC 6819 Colour-Magnitude Diagram
ngc_6819['absolute_magnitude'] = ngc_6819['phot_g_mean_mag'] - 5 * np.log10(ngc_6819['parallax']) + 5

plt.figure(figsize = (8, 6))
plt.scatter(ngc_6819['bp_rp'], ngc_6819['absolute_magnitude'], s = 10, c = ngc_6819['bp_rp'], cmap = 'coolwarm', marker = 'o')
plt.xlabel('Color (BP - RP)')
plt.ylabel('Absolute Magnitude (M)')
plt.gca().invert_yaxis() 
plt.title('Colour-Magnitude Diagram for NGC 6819 Cluster')
plt.show()
```

```python
# Filter the DataFrame to select rows where 'Cluster' == 'NGC_7789'
ngc_7789_cluster = filtered_ngc_7789[filtered_ngc_7789['Cluster'] == 'NGC_7789']

# Calculate the absolute magnitude for the selected data
ngc_7789_cluster['absolute_magnitude'] = ngc_7789_cluster['phot_g_mean_mag'] - 5 * np.log10(ngc_7789_cluster['parallax']) + 5

# Create the Color-Magnitude Diagram for NGC 7789 Cluster
plt.figure(figsize=(8, 6))
plt.scatter(ngc_7789_cluster['bp_rp'], ngc_7789_cluster['absolute_magnitude'], s=10, c=ngc_7789_cluster['bp_rp'], cmap='coolwarm', marker='o')
plt.xlabel('Color (BP - RP)')
plt.ylabel('Absolute Magnitude (M)')
plt.gca().invert_yaxis()
plt.title('Colour-Magnitude Diagram for NGC 7789 Cluster')
plt.show()
```
